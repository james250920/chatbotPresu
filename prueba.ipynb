```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options
import traceback
import time

# 1. Configurar Chrome
chrome_options = Options()
chrome_options.add_experimental_option("detach", True)

driver = webdriver.Chrome(options=chrome_options)

try:
    # 2. Abrir tu servidor local
    url = "http://127.0.0.1:5500/index.html"
    driver.get(url)
    
    wait = WebDriverWait(driver, 15)

    # 3. Primer Clic: Abrir el contenedor del chat
    print("Abriendo el menú del chat...")
    chat_button = wait.until(EC.element_to_be_clickable((By.ID, "chatButton")))
    chat_button.click()
    print("1. Menú del chat desplegado.")

    # 4. Saltar dentro del iframe
    print("Esperando a que cargue el iframe de Copilot Studio...")
    wait.until(EC.frame_to_be_available_and_switch_to_it((By.ID, "chatFrame")))
    print("¡Conectado con éxito al interior del iframe del Bot!")

    # 5. PASO CORRECTOR: Buscar el botón intermedio de inicio dentro del iframe
    # Copilot Studio suele poner un botón de bienvenida (fui-Button) antes de activar el input
    print("Buscando el botón intermedio de activación (fui-Button) dentro del iframe...")
    try:
        # Buscamos cualquier botón que contenga las clases de Fluent UI que mencionaste
        next_button_locator = "//button[contains(@class, 'fui-Button')]"
        next_button = wait.until(EC.element_to_be_clickable((By.XPATH, next_button_locator)))
        driver.execute_script("arguments[0].click();", next_button)
        print("-> Botón intermedio de Copilot presionado.")
        
        # Le damos 2 segundos a Microsoft para que procese el clic y dibuje la caja de texto
        time.sleep(2)
    except Exception:
        print("-> No se detectó botón intermedio o ya estaba activo, continuando directo a la caja de texto...")

    # 6. Escribir en la caja de texto de Copilot
    print("Buscando la caja de texto interna del chat...")
    # Cambiamos a presence_of_element_located por si el estilo CSS la oculta parcialmente al inicio
    chat_input_locator = (By.CLASS_NAME, "webchat__send-box-text-box__input")
    chat_input = wait.until(EC.presence_of_element_located(chat_input_locator))
    
    # Aseguramos el foco en el input mediante JavaScript y escribimos
    driver.execute_script("arguments[0].focus();", chat_input)
    chat_input.send_keys("Que es una certificación")
    print("2. Texto ingresado con éxito en el input.")

    # 7. Botón de enviar de Copilot (Web Chat)
    print("Enviando mensaje...")
    send_button_locator = "button[class*='webchat__send-button']"
    send_button = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, send_button_locator)))
    
    driver.execute_script("arguments[0].click();", send_button)
    print("3. ¡Mensaje enviado a Copilot Studio exitosamente!")

except Exception as e:
    print("\n❌ Error detectado en el flujo:")
    traceback.print_exc()
```
